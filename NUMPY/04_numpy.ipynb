```python
import numpy as np
```

# 📊 Common Data Types

## 🔢 Integers
- `int32` – 32-bit integer  
- `int64` – 64-bit integer  

## 🔢 Floating Point Numbers
- `float32` – 32-bit floating point number  
- `float64` – 64-bit floating point number  

## ✅ Boolean
- `bool` – Represents `True` or `False`  

## 🔣 Complex Numbers
- `complex64` – Complex number with 32-bit precision  
- `complex128` – Complex number with 64-bit precision  

## 🔤 Strings
- `S` – Byte string (byte-str)  
- `U` – Unicode string (unicode-str)  

## 📦 Object
- `object` – Generic Python objects  

```python
#Data Type 


arr = np.array([1,2,3,4,5])

print(arr,arr.dtype)

```

```python
#data Types

arr = np.array([3 + 5j])
print(arr,arr.dtype)
```

```python
arr1 = np.array([3 + 5j])
arr2 = np.array([2 + 3j])


print(arr1 + arr2)
print(arr1 - arr2)
```

```python
arr = np.array(["prime",{1,2,3},3.14])

print(arr,arr.dtype)
```

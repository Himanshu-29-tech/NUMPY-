```python
#create scratch
arr1 = np.zeros((2,3),dtype="int64") #prefill  
#we an also change type through dtype 
print(arr1,arr1.shape)

arr3 = np.ones((5,6))
print(arr3,arr3.shape)


arr4 = np.full((3,4),100) #prefill with value
print(arr4, arr4.shape)


arr5 = np.eye((3)) #prefill with identity matrix  pass single value.     identity matrix-only diagonal value = 1 
print(arr5, arr5.shape)

arr6 = np.arange(1,10) # range like a python
print(arr6, arr6.shape)


arr7 = np.linspace(1,100,4) #evenly spaced arrays       syntax:-👉linespace(start,end,total number)
print(arr7, arr7.shape)



```

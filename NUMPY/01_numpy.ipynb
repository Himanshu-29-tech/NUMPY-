```python
#execution performance
size = 1_000_000

#python list
py_list = list(range(size)) #[1,2,3,4..........]
start = time.time() #give current time value
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seeconds")


#numpay arrays
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"pytnon list time = {end-start} seconds")
```

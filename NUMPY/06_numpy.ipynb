```python
import numpy as np
```

```python
#Mathematical fux

#aggregation

arr =np.array([1,2,3,4,5])

print(np.sum(arr))      # Sum of all elements
print(np.prod(arr))     # Product (multiplication) of all elements
print(np.min(arr))      # Smallest value in array
print(np.max(arr))      # Largest value in array
print(np.argmin(arr))   # Index of smallest value
print(np.argmax(arr))   # Index of largest value
print(np.mean(arr))     # Average value
print(np.std(arr))      # Standard deviation (spread of data)
print(np.median(arr))   # Middle value after 
print(np.var(arr))      # Variance (measure of data spread, square of standard deviation)
```

**Variance = average of squared differences from mean**

```python
#power function

arr = np.array([1,2,3,4,5])


print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
#log

print(np.log(arr))
print(np.log2(arr))
print(np.exp(arr))
```

```python
#rounding

print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))

arr = np.array([1,2,-8,3,4,-5,2,-5])

print(np.unique(arr))
print(np.sqrt(arr))

print(np.abs(arr))
```
